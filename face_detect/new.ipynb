```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
model
```

```python
image = '../for_yolo/1596614532_710653.png'
```

```python
results = model(image)
```

```python
results
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
cap = cv2.VideoCapture('glina_dance.mp4')
while cap.isOpened():
    ret, frame = cap.read()

    results = model(frame)

    cv2.imshow('DANCING', np.squeeze(results.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cv2.release()
cv2.destroyAllWindows()
```
